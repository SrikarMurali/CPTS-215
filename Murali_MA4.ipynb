You can make a queue using these function by using the push and pop function for the enqueue and dequeue methods. To do this you must first implement the queue as a list, and then when adding items to the queue just push them in at the front. When trying to dequeue items, just traverse the list with a loop until you get to the oldest item in the queue. You could use two stacks for this. To enqueue you should follow these steps. While stack1 is not empty, push everything from satck1 to stack2. Push x to stack1 (assuming size of stacks is unlimited). Push everything back to stack1. To dequeue follow these steps. If stack1 is empty then error. If not then pop an item from stack1 and return it.

#2

The worst case run time will be $O(n)$ when either using enqueue or dequeue depending on how you set up the queue.

#3

The pop operation will perform n dequeues for each of the enqueues. However there will probably be around $n^2$ pops.


```python
##Queue List implementation
import random
import timeit
import time
import urllib.request

class Queue:
    def __init__(self):
        self.items = []
        self.dqCount = 0
    
    def isEmpty(self):
        return self.items == []
    
    def enqueue(self, item):
        self.items.insert(0, item)
    
    def dequeue(self):
        self.dqCount+=1
        return self.items.pop()
    
    def size(self):
        return len(self.items)
    
def main():
    n = 10000
    d = Queue()
    data = urllib.request.urlopen('https://raw.githubusercontent.com/gsprint23/cpts215/master/microassignments/files/words.txt').read()

    for i in range(len(data)):
        d.enqueue(data[i])

    start = time.time()
    
    while d.isEmpty() == False:
        d.dequeue()
    end = time.time()
    
    print(end - start)
    print('Dequeue Count', d.dqCount)

if __name__ == '__main__':
    main()
```

```python
#Queue Implementation using two stacks
#Code for question 2 and 3

import timeit
import time
import urllib.request

class Stack():

    def __init__(self):
        self.stk = []

    def pop(self):
        """raises IndexError if you pop when it's empty"""
        return self.stk.pop()

    def push(self, elt):
        self.stk.append(elt)

    def is_empty(self):
        return len(self.stk) == 0

    def peek(self):
        if not self.stk.is_empty():
            return self.stk[-1]


class Queue():

    def __init__(self):
        self.q = Stack()  
        self.b = Stack()  
        self.front = None
        self.size = 0
        self.popCount = 0

    def is_empty(self):
        return self.q.is_empty()

    def peek(self):
        if self.q.is_empty():
            return None
        else:
            return self.front

    def enqueue(self, elt):
        self.front = elt
        self.q.push(elt)
        self.size+=1

    def dequeue(self):
        """raises IndexError if you dequeue from an empty queue"""
        while not self.q.is_empty() > 0:
            elt = self.q.pop()
            self.b.push(elt)
            self.popCount+=1
        val = self.b.pop()
        self.popCount+=1
        elt = None
        while not self.b.is_empty() > 0:
            elt = self.b.pop()
            self.popCount+=1
            self.q.push(elt)
        self.front = elt
        self.size-=1
        return val
    
    
def main():
    c = Queue()
    n = 100
    #for i in range(n):
     #   c.enqueue(i)
    start = time.time()
    data = urllib.request.urlopen('https://raw.githubusercontent.com/gsprint23/cpts215/master/microassignments/files/words.txt').read()


    for i in range(len(data)):
        c.enqueue(data[i])
    while c.is_empty() == False:
        c.dequeue()
    end = time.time()
    print(end - start)
    print("Pop count", c.popCount)
if __name__ == '__main__':
    main()
```

```python
##Linked list implementation of Stacks/Queues

import time
import urllib.request

class Node:
    def __init__(self, initdata):
        self.data = initdata
        self.next = None
        
    def getData(self):
        return self.data
    
    def getNext(self):
        return self.next
    
    def setData(self, newdata):
        self.data = newdata
        
    def setNext(self, newnext):
        self.next = newnext
        
class Queue:
    def __init__(self):
        self.front = None
        self.back = None
        self.numlinks = 0
        self.dqCount = 0
        
    def enqueue(self, data):
        newNode = Node(data)
        
        if self.front is None:
            self.front = newNode
            self.back = self.front
        else:
            self.back.setNext(newNode)
            self.back = self.back.getNext()
        self.numlinks+=1
    
    def dequeue(self):
        if self.isEmpty():
            print("Empty Queue")
            return None
        frontNode = self.front
        if self.front.getNext() is None:
            self.front = None
            self.back = None
        else:
            self.front = self.front.getNext()
        self.numlinks-=1
        self.dqCount+=1
        return frontNode
    
    def qprint(self):
        if not self.isEmpty():
            frontNode = self.front
            while not frontNode is None:
                print(frontNode.getData(), end = " -> ")
                frontNode = frontNode.getNext()
            print("null")
        else:
            print("Empty Queue")
            
    def isEmpty(self):
        return self.numlinks == 0
    
    def size(self):
        return self.numlinks
    
class Stack:
    
    def __init__(self):
        self.head = None
        self.numlinks = 0
    
    def __str__(self):
        if self.isEmpty():
            print("Empty Stack")
        frontNode = self.head
        while frontNode is not None:
            print(frontNode.getData(), end = " -> ")
            frontNode = frontNode.getNext()
        print("null")
    
    def push(self, item):
        newNode = Node(item)
        newNode.next = self.head
        self.head = newNode
        self.numlinks+=1
    
    def pop(self):
        if self.isEmpty():
            print("Stack is Empty")
            return None
        topNode = self.head
        self.head = self.head.getNext()
        topNode.next = None
        self.numlinks-=1
        return topNode
    
    def isEmpty(self):
        return self.numlinks == 0
    
    def size(self):
        return self.numlinks

def main():
    q = Queue()
    n = 10000
    data = urllib.request.urlopen('https://raw.githubusercontent.com/gsprint23/cpts215/master/microassignments/files/words.txt').read()


    for i in range(len(data)):
        q.enqueue(data[i])
    start = time.time()
    while q.isEmpty() == False:
        q.dequeue()
    end = time.time()
    
    print(end - start)
    print('Dequeue Count', q.dqCount)
if __name__ == '__main__':
    main()
    
```
